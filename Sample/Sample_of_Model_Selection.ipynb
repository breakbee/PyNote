# Sample of Model Selection

ベイズ情報量規準(BIC)によるモデル選択の例  
ガウス混合モデルのハイパーパラメータ(=クラスタの個数)を選択する  

```python
import numpy as np
import math

from scipy.stats import multivariate_normal

###  サンプルの生成
# 2次元2混合正規分布から1000サンプル生成

dim = 2
mean1 = np.zeros(dim)+10
mean2 = np.zeros(dim)-10
cov1 = np.eye(dim)
cov2 = 4 * np.eye(dim)
num1 = 500
num2 = 500

samples = np.vstack((
	np.random.multivariate_normal(mean1, cov1, num1),
	np.random.multivariate_normal(mean2, cov2, num2)
	))
```

```python
### EMアルゴリズムでガウス混合モデルで学習し、BICを計算

from sklearn import mixture

cluster_nums = range(1, 10)
bic_results = []

for i, k in enumerate(cluster_nums):
	gmm = mixture.GMM(n_components=k, n_iter=10000, covariance_type='full')
	gmm.fit(samples)
	bic_results.append(gmm.bic(samples))
```

```python
### 可視化

import matplotlib.pyplot as plt

plt.bar(cluster_nums, bic_results, align='center')
plt.show()
```

```python
### 可視化(その2)

import pylab as pl

pl.plot(cluster_nums, bic_results)
pl.show()
```

真の分布の混合数である2が選択されていることが確認できた  

ベイズ情報量規準にはモデル選択の一致性(この例で言えば、サンプルが多ければ正しい混合数を選べる)がある  
ベイズ情報量規準は自由エネルギーと呼ばれる値を最小化する  
なお、変分ベイズ法によるガウス混合モデルの学習も自由エネルギーを最小化するので、BICと等価である
