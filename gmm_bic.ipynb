```python
import numpy as np
import pylab as pl
import math

from sklearn import mixture

from scipy.stats import multivariate_normal

### BICによるGMMのモデル選択

dim = 2
mean1 = np.zeros(dim)+10
mean2 = np.zeros(dim)-10
cov1 = np.eye(dim)
cov2 = 4 * np.eye(dim)
num1 = 500
num2 = 500

samples = np.vstack((
	np.random.multivariate_normal(mean1, cov1, num1),
	np.random.multivariate_normal(mean2, cov2, num2)
	))

def pdf(x, means, precs, pi):
	return sum([pi[i]*multivariate_normal.pdf(x, means[i], np.linalg.inv(precs[i])) for i in range(pi.shape[0])])

cluster_nums = range(1, 10)
bic_results = np.zeros(len(cluster_nums))
gmm_list = []
for i, k in enumerate(cluster_nums):
	gmm = mixture.GMM(n_components=k, n_iter=10000, covariance_type='full')
	gmm.fit(samples)
	bic_results[i] = gmm.bic(samples)

	gmm_list.append(gmm)

	#bic_results[i] = -2 * sum(gmm.score(samples)) + k * math.log(samples.shape[0])
	# VBGMMのscoreとbicにバグがある?

	# log_score = []
	# for sample in samples:
	# 	p = pdf(sample, gmm.means_, gmm.precs_, gmm.weights_)
	# 	log_score.append(math.log(p))
	# bic_results[i] = -2 * sum(log_score) + k * math.log(samples.shape[0])

	# 精度行列の値がおかしい? -> 単位行列分足されてる?
	# _get_covars() で共分散行列取れる
	# というか推定自体おかしい
	# 実装バグってる気がする

	# サンプルしてみよう！
	
pl.plot(cluster_nums, bic_results)
pl.show()
```
