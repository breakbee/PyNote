# Stopword Cheatsheet

ストップワードに困ったらとりあえずこれを試そう  

- **日本語ストップワード**  
http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt  

- **英語ストップワード**  
http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/English.txt  

```python
### 日本語ストップワード
stopword_ja = set()

with open('stopword/Japanese.txt', 'r', encoding='utf-8') as f:
	for line in f:
		word = line.rstrip('\n')
		stopword_ja.add(word)
```

```python
for word in stopword_ja:
    print(word, end=' ')
```

```python
### 英語ストップワード
stopword_en = set()

with open('stopword/English.txt', 'r', encoding='utf-8') as f:
	for line in f:
		word = line.rstrip('\n')
		stopword_en.add(word)
```

```python
for word in stopword_en:
    print(word, end=' ')
```
