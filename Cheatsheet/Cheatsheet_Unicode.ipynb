# Unicode Cheatsheet

**PythonでUnicode文字を扱う時のためのチートシート**

```python
# 最初と最後の文字が分かるならこれ
for c in range(ord('ぁ'), ord('ん')):
	print(chr(c), end=" ")
```

**ユニコード表**  

半角英数字と記号：0x0020～0x007e  
ひらがな：0x3041～0x3093  
全角カナ：0x30a1～0x30f6  
半角カナ：0xff61～0xffa0  
全角英数字：0xff10～0xff5a  
全角空白：0x3000  
半角空白：0x0020

```python
# ひらがな
for c in range(int('0x3041', 16), int('0x3093', 16)):
	print(chr(c), end=" ")
```

```python
# ひらがなと全角記号
for c in range(int('0x3000', 16), int('0x30FF', 16)+1):
	print(chr(c), end=" ")
```

```python
# 半角英数字と記号
for c in range(int('0x0020', 16), int('0x007e', 16)+1):
	print(chr(c), end=" ")
```

```python
# 全角カナ
for c in range(int('0x30a1', 16), int('0x30f6', 16)+1):
	print(chr(c), end=" ")
```

```python
# 半角カナ
for c in range(int('0xff61', 16), int('0xffa0', 16)+1):
	print(chr(c), end=" ")
```

```python
# 全角英数字
for c in range(int('0xff10', 16), int('0xff5a', 16)+1):
	print(chr(c), end=" ")
```
