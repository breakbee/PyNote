# Implementation of k-Means

k-Means 法の実装とサンプルです  

```python
import numpy as np
import random

### k-Means の実装
class KMeans:
	def __init__(self, k):
		# クラスタ数
		self.k = k

    # クラスタリング
	def clustering(self, samples):
		k = self.k
		n, dim = samples.shape

		# 初期化
		means = self.__initial_means(samples, k)
		clusters = np.zeros(n) - 1

		while True:
			# クラスタの割り当ての更新
			new_clusters = self.__update_clusters(samples, means)
			# 各クラスタの平均の更新
			new_menas = self.__update_means(samples, new_clusters, k)
			# 収束するまで更新
			dif = new_clusters - clusters
			if np.sum(dif) == 0:
				break
			means = new_menas
			clusters = new_clusters

		return means, clusters

	# 初期値の割り当て
	def __initial_means(self, samples, k):
		return random.sample(list(samples), k)

	# 1つのサンプルについてクラスタの割り当ての更新
	def __update_cluster(self, sample, means):
		dists = [self.__square_dist(sample, mean) for mean in means]
		min_dist = min(dists)
		index = dists.index(min_dist)
		return index

	# サンプル全体について各クラスタの平均の更新
	def __update_clusters(self, samples, means):
		clusters = [self.__update_cluster(sample, means) for sample in samples]
		return np.array(clusters)

	# 各クラスタの平均の更新
	def __update_means(self, samples, clusters, k):
		n, dim = samples.shape
		means = np.zeros([k, dim])
		nums = np.zeros(k)
		for i in range(n):
			index = clusters[i]
			means[index] += samples[i]
			nums[index] += 1
		return means * np.transpose(np.tile(1.0/nums, (k, 1)))

	# 距離の計算
	def __square_dist(self, x1, x2):
		dif = x1 - x2;
		return np.dot(dif, dif)
```

2つの2次元正規分布から生成したサンプルを用いてテスト

```python
# サンプルの生成
dim = 2
mean1 = np.zeros(dim)+10
mean2 = np.zeros(dim)-10
cov1 = np.eye(dim)
cov2 = 5 * np.eye(dim)
num1 = 1000
num2 = 500

samples = np.vstack((
	np.random.multivariate_normal(mean1, cov1, num1),
	np.random.multivariate_normal(mean2, cov2, num2)
	))
```

```python
# クラスタリング
km = KMeans(k=2)
means, clusters = km.clustering(samples)
```
