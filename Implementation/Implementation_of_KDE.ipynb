# Implementation of KDE

**カーネル密度推定の実装とサンプル**  

カーネル密度推定はヒストグラム法の拡張である。  
ヒストグラムの平滑化としてガウス関数を用いている。  

- ヒストグラムのビン幅を用いたカーネルを sphere kernel  
- 多変量ガウス分布として共分散行列を用いたカーネルを unconstrained kernel  

とする。

```python
import numpy as np
import math

class KDE:

    # sphere kernelによる推定
	def estimate_s(self, x, samples, h):
		n, dim = samples.shape
		p = sum([self.__sphere_kernel((x-sample)/h) for sample in samples])
		p /= n*(h**dim)
		return p

    # unconstrained kernel による推定
	def estimate_u(self, x, samples, H):
		p = sum([self.__unconstrained_kernel(x, sample, H) for sample in samples])
		n = samples.shape[0]
		p /= n
		return p

    # sphere kernel
	def __sphere_kernel(self, x):
		dim = x.size
		tmp = math.exp(-np.dot(x, x)/2)
		value = tmp / (2*np.pi)**(float(dim)/2)
		return value

    # unconstrained kernel 
	def __unconstrained_kernel(self, x, mean, cov):
		det = np.linalg.det(2*np.pi*cov)
		tmp = np.dot(np.linalg.inv(cov), x-mean)
		tmp = np.dot(x-mean, tmp)	
		tmp = -tmp/2.0
		value = math.exp(tmp)/math.sqrt(det)
		return value
```

以下、2次元標準正規分布を用いた実験を行う。

```python
# サンプルの生成
dim = 2
mean = np.zeros(dim)
cov = np.eye(dim)
num = 1000
samples = np.random.multivariate_normal(mean, cov, num)

samples.shape
```

```python
# KDEのパラメータ設定
# 厳密には交差検定法による決定が必要

# sphere kernel のビン幅
h = 0.25

# unconstrained kernel の共分散行列
H = np.cov(np.transpose(samples)) * 0.1

kde = KDE()
```

```python
# 3次元描画用のユーティリティ関数
def draw3d(xs, ys, xe, ye, div, func):
	xlist = np.arange(xs, xe+div, div)
	ylist = np.arange(ys, ye+div, div)
	X, Y = np.meshgrid(xlist, ylist)
	P = [[X[i][j], Y[i][j]] for i in range(xlist.size) for j in range(ylist.size)]
	Z = np.asarray([func(point) for point in P]).reshape(xlist.size, ylist.size)
	return [X, Y, Z]
```

```python
# 描画
from scipy.stats import multivariate_normal

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X1, Y1, Z1 = draw3d(-3, -3, 3, 3, 0.25, lambda x : kde.estimate_u(x, samples, H))
X2, Y2, Z2 = draw3d(-3, -3, 3, 3, 0.25, lambda x : kde.estimate_s(x, samples, h))
Xt, Yt, Zt = draw3d(-3, -3, 3, 3, 0.25, lambda x : multivariate_normal.pdf(x, mean, cov))

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X1, Y1, Z1, color="blue")
ax.plot_wireframe(X2, Y2, Z2, color="green")
ax.plot_wireframe(Xt, Yt, Zt, color="red")
plt.show()
```

赤が真の分布  
青が unconstrained kernel による推定  
緑が sphere kernel による推定
