# Regular Expression Cheatsheet

**Python で正規表現の例**  

```python
import re

text = 'Python, Java, C++, etc... と、プログラミング言語には様々な種類があり、それぞれの使い道がある。奥が深い。'
```

**一番簡単な例**  
**半角文字以外を抽出する正規表現**  

**参考**  
正規表現サンプル(全角文字を検索する)  
http://hodade.adam.ne.jp/seiki/page.php?s_zenkaku

```python
regexp1 = r'[^\x01-\x7E]+'

pattern = re.compile(regexp1)
for word in pattern.findall(text):
    print(word)
```

日本語をしっかり指定する例  
この例だとマッチし損ねてる単語がある  

**参考**  

Python による日本語自然言語処理  
https://nltk.googlecode.com/svn/trunk/doc/book-jp/ch12.html

```python
# (,.、。)などの扱いも考えること
# 漢字の範囲をしっかり
# この例だと「々」がマッチできていない
regexp2 = r'[ぁ-んァ-ンー\u4e00-\u9FFF]+'

pattern = re.compile(regexp2)
for word in pattern.findall(text):
    print(word)
```
