# Shell Command Cheatsheet

**Python からシェルコマンドを実行するには subprocess モジュールを使用する**  

````python
import subprocess

subprocess.call(["cat", "README.md"])
````

コマンドが完了したら戻り値がかえってくる(この場合は0)  
出力は以下  

```
PyNote
======
My iPython Notebook for Machine Learning and Statistics.

see : http://nbviewer.ipython.org/github/breakbee/PyNote/tree/master/
```

**パイプ処理等面倒な処理を扱う場合にはシェルスクリプトをかますと楽**  

例えば cat の出力を cat.txt に保存したいときは, 次のようなシェルスクリプト cat.sh を書き, sh コマンドで実行する

```
#!/bin/sh
cat $1 > $2
```

````python
commands = ["sh", "cat.sh", "README.md" ,"cat.txt"]
subprocess.call(commands)
````

cat.txt の中身は同上
